## GDP Regression Dataset

#### Load the country GDP and Education spend dataset

```python
gdp_df  = pd.read_csv('https://raw.githubusercontent.com/magtanggol03/dec130/master/datasets/economic-indicator/gdp.csv', header = 2)
gov_exp_df = pd.read_csv('https://raw.githubusercontent.com/magtanggol03/dec130/master/datasets/economic-indicator/gov_exp.csv', header = 2)
```

#### Only get 2013 data, merge the data in one data frame. Drop all countries that have missing data

```python
merged_df = gov_exp_df[['Country Name', '2013']].merge(gdp_df[['Country Name', '2013']], on = 'Country Name').dropna() #get only 2013, merge and drop
merged_df = merged_df.set_index('Country Name')#Set dataframe index to countries
merged_df.columns = ['Spending on Education(%)', 'GDP Growth (%)'] #Set proper column titles
merged_df = merged_df.drop(['Timor-Leste','South Sudan', 'Sierra Leone'],axis = 0)

merged_df.head()
```

#### Get Distribution of Education Spend %

```python
f, ax = plt.subplots(1, 2, figsize = (10,10), gridspec_kw = {'width_ratios':[1, 3]}) #Create subplots to allow multiple plots in one
merged_df.iloc[:, 0].hist(color = col[0], ax = ax[1], orientation='horizontal', bins = 15) #plot histogram
merged_df.iloc[:, 0].plot(kind = 'box', ax = ax[0], color = col[0]) #plot boxplot
plt.title('Distribution of 2013 Education Spend (%)') #set title
plt.show()
```

#### Get Distribution of GDP Growth (%)

```python
f, ax = plt.subplots(1, 2, figsize = (10,10), gridspec_kw = {'width_ratios':[1, 3]}) #Create subplots to allow multiple plots in one
merged_df.iloc[:, 1].hist(color = col[2], ax = ax[1], orientation='horizontal', bins = 10)  #plot histogram
merged_df.iloc[:, 1].plot(kind = 'box', ax = ax[0], color = col[1]) #plot boxplot
plt.title('Distribution of 2013 GDP Growth (%)') #set title
plt.show()
```

#### Plot scatter plot of GDP and Education Spend

```python
plt.figure(figsize = (10,10))
plt.scatter(x = merged_df['Spending on Education(%)'], y = merged_df['GDP Growth (%)'], color = col[0])
plt.xlabel(merged_df.columns[0])
plt.ylabel(merged_df.columns[1])
plt.title('2013 Government Spending on Education (%) vs GDP Growth (%) per Country')
plt.show()
```

```python
merged_df.corr() #Get correlation
```

#### Fit our Ordinary Least Squares (OLS) Regression Model with our data
Note that we transform our X to allow for a constant variable

```python
model = sm.OLS(merged_df['GDP Growth (%)'], sm.add_constant(merged_df['Spending on Education(%)'])).fit()
model.summary()
```

#### Plot the resulting fitted line with our actual data

```python
const = model.params['const']
beta = model.params['Spending on Education(%)']
plt.figure(figsize = (10,10))
plt.scatter(x = merged_df['Spending on Education(%)'], y = merged_df['GDP Growth (%)'], color = col[0])
plt.xlabel(merged_df.columns[0])
plt.ylabel(merged_df.columns[1])
plt.title('2013 Government Spending on Education (%) vs GDP Growth (%) per Country')
pylab.plot(merged_df['Spending on Education(%)'], const + beta*merged_df['Spending on Education(%)'], color = col[2])
plt.show()
```

#### Compare result with standardized values

```python
meansd = {'mean': merged_df.mean(), 'sd':  merged_df.std()}

std_merge = merged_df.copy()
std_merge['Spending on Education(%)'] = (merged_df['Spending on Education(%)'] - meansd['mean']['Spending on Education(%)'])/meansd['sd']['Spending on Education(%)']
std_merge['GDP Growth (%)'] = (merged_df['GDP Growth (%)'] - meansd['mean']['GDP Growth (%)'])/meansd['sd']['GDP Growth (%)']

model2 = sm.OLS(std_merge['GDP Growth (%)'], sm.add_constant(std_merge['Spending on Education(%)'])).fit()
model2.summary()
```

#### Checking Model Assumptions

```python
fig, ax = plt.subplots(2, 1, figsize = (10, 5))
residuals = merged_df['GDP Growth (%)'].values - model.predict() #Calculate residuals

# Make residual plot
ax[0].scatter(x = merged_df['Spending on Education(%)'], y= residuals , color = col[1])
ax[0].set_title('Residual Plot')
ax[0].set_xlabel('Spending on Education(%)')
ax[0].set_ylabel('Residuals')

#Make QQ Plot
stats.probplot(residuals, dist="norm", plot= ax[1])
ax[1].get_lines()[0].set_markerfacecolor(col[2])
ax[1].get_lines()[1].set_color(col[4])
ax[1].set_title('QQ-Plot')
plt.tight_layout()
plt.show()
```

## HEADHUNTING DATASET

#### Load headhunting dataset

```python
headhunt_df = pd.read_csv('https://raw.githubusercontent.com/magtanggol03/dec130/master/datasets/headhunting.csv', header = 0)
headhunt_df.head()
```

#### Run regression on Days it takes to hire vs Company Size

```python
hh_model = sm.OLS(headhunt_df['DAYS'], sm.add_constant(headhunt_df.drop(['DAYS', 'Position'], axis = 1))).fit()
hh_model.summary()
```

#### Plot fitted line versus scatter plot

```python
plt.figure(figsize = (10, 6)) #set size of canvas
plt.scatter(x =headhunt_df.SIZE, y = headhunt_df.DAYS, color = col[2]) #Draw scatter
y_est = headhunt_df.SIZE*hh_model.params['SIZE'] + hh_model.params['const'] #get fitted Y from the model parameters
x = headhunt_df.SIZE# get X
pylab.plot(x, y_est, color = col[4])

#Set plot labels
plt.xlabel('Size of Company')
plt.ylabel('Days to Hire')
plt.title('Days to Hire vs Size of Company')
plt.show()
```

```python
plt.figure(figsize = (10, 6))

#Plot scatter plot with different colors per position
plt.scatter(x =headhunt_df[headhunt_df.Position == 'C-Suite'].SIZE, y = headhunt_df[headhunt_df.Position == 'C-Suite'].DAYS,  color = col[2], label = 'C-Suite')
plt.scatter(x =headhunt_df[headhunt_df.Position != 'C-Suite'].SIZE, y = headhunt_df[headhunt_df.Position != 'C-Suite'].DAYS,  color = col[3], label = 'Non C-Suite')

#Get fitted line
y_est = headhunt_df.SIZE*hh_model.params['SIZE'] + hh_model.params['const']
x = headhunt_df.SIZE

#Plot rest of the plot
pylab.plot(x, y_est, color = col[4], label = 'Fitted Line')
plt.xlabel('Size of Company')
plt.ylabel('Days to Hire')
plt.title('Days to Hire vs Size of Company')
plt.legend()
plt.show()
```

#### Convert Yes/No to Binary

```python
def convCategorical(x):
    #User defined function to convert categorical to binary 
    if x == 'C-Suite':
        return 1
    else:
        return 0
    
headhunt_df.Position = headhunt_df.Position.apply(convCategorical)
headhunt_df.head()
```

#### Add Position as a factor

```python
#Fit the new model
hh_model = sm.OLS(headhunt_df['DAYS'], sm.add_constant(headhunt_df.drop('DAYS', axis = 1))).fit()
hh_model.summary()

plt.figure(figsize = (10, 6))

#Plot scatter plots
plt.scatter(x =headhunt_df[headhunt_df.Position == 1].SIZE, y = headhunt_df[headhunt_df.Position == 1].DAYS,  color = col[2], label = 'C-Suite')
plt.scatter(x =headhunt_df[headhunt_df.Position != 1].SIZE, y = headhunt_df[headhunt_df.Position != 1].DAYS,  color = col[3], label = 'Non C-Suite')

#Plot fitted line for Position = 0 
y_est = 0*hh_model.params['Position'] + headhunt_df.SIZE*hh_model.params['SIZE'] + hh_model.params['const']
x= headhunt_df.SIZE
pylab.plot(x, y_est, color = col[4], label = 'Fitted Line when Not a CEO')


#Plot fitted line for Position = 1
y_est = 1*hh_model.params['Position'] + headhunt_df.SIZE*hh_model.params['SIZE'] + hh_model.params['const']
x = headhunt_df.SIZE
pylab.plot(x, y_est, color = col[5], label = 'Fitted Line when a CEO')

#Plot rest of the plot
plt.xlabel('Size of Company')
plt.ylabel('Days to Hire')
plt.title('Days to Hire vs Size of Company')
plt.legend()
plt.show()
```

#### Add interaction terms

```python
headhunt_df['interaction'] = headhunt_df['Position']*headhunt_df['SIZE']  #Add interaction terms
headhunt_df.loc[[0,20,40,60,80,95]]
```

```python
hh_model = sm.OLS(headhunt_df['DAYS'], sm.add_constant(headhunt_df.drop('DAYS', axis = 1))).fit()
hh_model.summary()
```

```python
#Fit the model
hh_model = sm.OLS(headhunt_df['DAYS'], sm.add_constant(headhunt_df.drop('DAYS', axis = 1))).fit()
hh_model.summary()

plt.figure(figsize = (10, 6))
#Plot the scatter plots
plt.scatter(x =headhunt_df[headhunt_df.Position == 1].SIZE, y = headhunt_df[headhunt_df.Position == 1].DAYS,  color = col[2], label = 'C-Suite')
plt.scatter(x =headhunt_df[headhunt_df.Position != 1].SIZE, y = headhunt_df[headhunt_df.Position != 1].DAYS,  color = col[3], label = 'Non C-Suite')

#Plot the fitted lines
x = headhunt_df.SIZE
y_est = 0*hh_model.params['interaction']*headhunt_df.SIZE*headhunt_df.Position + 0*hh_model.params['Position'] + headhunt_df.SIZE*hh_model.params['SIZE'] + hh_model.params['const']

pylab.plot(x, y_est, color = col[4], label = 'Fitted Line when Not a CEO')

y_est = 1*hh_model.params['interaction'] *headhunt_df.SIZE*1 + 1*hh_model.params['Position'] + headhunt_df.SIZE*hh_model.params['SIZE'] + hh_model.params['const']

pylab.plot(x, y_est, color = col[5], label = 'Fitted Line when a CEO', lw = 1)

#Plot the rest of the plot
plt.xlabel('Size of Company')
plt.ylabel('Days to Hire')
plt.title('Days to Hire vs Size of Company')
plt.legend()
plt.show()
```

## TELEMARKETING DATA SET

#### Load dataset

```python
tele_df = pd.read_csv('https://raw.githubusercontent.com/magtanggol03/dec130/master/datasets/tele.csv', header = 0).drop('Unnamed: 0', axis = 1)
tele_df.head()
```

```python
tele_df.plot.scatter('Months', 'Calls', figsize = (10, 5))
plt.title('Number of Calls vs Months Working')
plt.show()
```

#### Fit a regular Linear regression model

```python
tele_model = sm.OLS(tele_df['Calls'], sm.add_constant(tele_df['Months'])).fit()
tele_model.summary()
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(x =tele_df.Months, y = tele_df.Calls, color = col[2])
pylab.plot(tele_df.Months, tele_df.Months*tele_model.params['Months'] + tele_model.params['const'], color = col[4])
plt.xlabel('Months')
plt.ylabel('Calls')
plt.title('Number of Calls vs Months Working')
plt.show()
```

#### Get a higher order term (squared)

```python
tele_df['Months2'] = tele_df.Months**2
tele_df['Months'] = tele_df.Months - tele_df.Months.mean()
tele_df['Months2'] = tele_df.Months2 - tele_df.Months2.mean()
tele_df.head()
```

```python
tele_model = sm.OLS(tele_df['Calls'], sm.add_constant(tele_df[['Months', 'Months2']])).fit()
tele_model.summary()
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(x =tele_df.Months, y = tele_df.Calls, color = col[2])
pylab.plot(tele_df.Months, tele_df.Months*tele_model.params['Months'] + tele_df.Months2*tele_model.params['Months2'] + tele_model.params['const'], color = col[4])
plt.xlabel('Months')
plt.ylabel('Calls')
plt.title('Number of Calls vs Months Working')
plt.show()
```

#### Converting Categorical to Dummy Variables

```python
churn_df = pd.read_csv("https://raw.githubusercontent.com/magtanggol03/dec130/master/datasets/telco-churn/test.csv")
churn_df.head()
```

#### pd.get_dummies(Series) creates one column for every category.
#### However, we need n-1 columns, so choose a "base" category and drop that from the columns.
So if all the dummy variables are 0, it refers to the base category

#### After getting the dummy variables, just use <span color="color:#ef4631">pd.concat([original data frame, dummy variables]).drop(original column, axis = 1)</span> to combine the new columns and remove the original.

```python
base = 'No'
pd.concat([churn_df['InternetService'], pd.get_dummies(churn_df['InternetService']).drop(base, axis = 1)], axis = 1)
```
